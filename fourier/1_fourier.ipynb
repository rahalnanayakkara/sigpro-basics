# Fourier Transform

```python
import numpy as np
import matplotlib.pyplot as plt
```

<center>

$ X(j\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} dt$

</center>

```python
T = 50
T_points = 10000

t = np.linspace(0, T, T_points).reshape(1,-1)
t.shape
```

#### Generate Monotone Signal

```python
freq = 3
omega = 2*np.pi*freq

amp = 1

phase = 0

x_t = amp*np.cos(omega*t+phase)
x_t.shape
```

```python
plt.figure(figsize=(12,5))
plt.plot(t.T, x_t.T)            # plot command needs column matrix or 1D vector
plt.xlabel("Time/s")
plt.ylabel("Amplitude")
plt.show()

print(f'Angular Frequency = {omega} rad/s')
```

```python
W = 50
W_points = 1000

w = np.linspace(0,W,W_points).reshape(-1,1)
w.shape
```

```python
comp_exp = np.exp(1j*np.dot(w,t))
comp_exp.shape
```

Each row in "comp_exp" matrix corresponds to a single frequency

```python
X_w = np.trapz(x_t*comp_exp, axis=1)

plt.plot(w, np.abs(X_w))
plt.xlabel("Frequency / (rad/s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
f = w/(2*np.pi)
plt.plot(f, np.abs(X_w))
plt.xlabel("Frequency / Hz")
plt.ylabel("Amplitude")
plt.show()
```

### Multi Tone Signal

```python
N_Tones = 3

tones = np.random.rand(N_Tones)*W/2
amps = np.random.rand(N_Tones)*amp

x_t = 0*t

plt.figure(figsize=(12,5))

for i in range(N_Tones):
    temp = amps[i]*np.cos(tones[i]*t+phase)
    x_t += temp
    plt.plot(t.T, temp.T)
    plt.xlabel("Time/s")
    plt.ylabel("Amplitude")
    print(f'Freq = {tones[i]}, Amp = {amps[i]} ')
```

```python
plt.figure(figsize=(12,5))
plt.plot(t.T, x_t.T)            # plot command needs column matrix or 1D vector
plt.xlabel("Time/s")
plt.ylabel("Amplitude")
plt.show()
```

```python
X_w = np.trapz(x_t*comp_exp, axis=1)

plt.plot(w, np.abs(X_w))
plt.xlabel("Frequency / (rad/s)")
plt.ylabel("Amplitude")
plt.show()
```

```python
f = w/(2*np.pi)
plt.plot(f, np.abs(X_w))
plt.xlabel("Frequency / Hz")
plt.ylabel("Amplitude")
plt.show()
```
